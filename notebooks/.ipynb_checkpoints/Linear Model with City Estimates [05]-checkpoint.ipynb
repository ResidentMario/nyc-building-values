Now let's build a second linear model, this one taking the city's estimates into account. How much better is our performance when we can take advantage of this previous work? What is the effect on variables of interest? Let's find out...

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
train_test = pd.read_csv("../data/processed/likely_sales_featurized.csv", index_col=0)
```

```python
nonassessment_features = list(set(train_test.columns) - {'Borough', 'Block', 'Lot', 'SalePrice', 'ZipCode', 'Address', 
                                                         'MarketValueSqFt'})
target_variable = 'MarketValueSqFt'
```

```python
nonassessment_train_test = train_test[nonassessment_features]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
y = train_test['MarketValueSqFt']
clf.fit(nonassessment_train_test, y)
```

```python
y_pred = clf.predict(nonassessment_train_test)
```

```python
import sklearn.metrics
```

```python
mse = sklearn.metrics.mean_squared_error(y, y_pred)
mse
```

```python
root_mse = mse**(1/2)
root_mse
```

```python
_*1200
```

Remember that our original model *without* city value estimation data resulted in an error a little over 100,000, meaning that adding city data resulted in an approximately 25% improvement in our mean error. That's super!

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
sns.kdeplot(y)
ax = plt.gca()
ax.set_xlim([0, 600])
plt.axvline(y.median(), c='black')
ax.add_patch(
    patches.Rectangle((y.median() - root_mse, 0), 2*root_mse, 0.006,
                      color='r', alpha=0.2)
)
```

```python
r_squared = sklearn.metrics.r2_score(y, y_pred)
r_squared
```

$R^2$ is now ~80%. This is great for such a simple classifier!

```python
coefs = list(zip(clf.coef_, nonassessment_features))
sorted(coefs)[::-1]
```

Again let's look at variable results.

```python
firecomps = [(a, b) for a, b in coefs if "FireComp" in b]
firecomps = pd.Series(data=[f[0]*1200 for f in firecomps], index=[f[1] for f in firecomps])
firecomps.sort_values(ascending=False).head(20).plot(kind='bar', figsize=(24, 10), fontsize=26)
```

FireComp (geography) has roughly the same importance.

```python
zoning = [(a, b) for a, b in coefs if "Zoning" in b]
zoning = pd.Series(data=[f[0]*1200 for f in zoning], index=[f[1] for f in zoning])
zoning.sort_values(ascending=False).plot(kind='bar', figsize=(24, 10), fontsize=20)
```

Zoning has flattened a little bit, but it's still pretty similar to the earlier model.

```python
owners = [(a, b) for a, b in coefs if "OwnerType" in b]
owners = pd.Series(data=[f[0]*1200 for f in owners], index=[f[1] for f in owners])
owners.sort_values(ascending=False).plot(kind='bar', figsize=(24, 10), fontsize=20)
```

The effect of ownership, on the other hand, has been *strongly* rescaled against Public ownership.

```python
prox = [(a, b) for a, b in coefs if "ProxCode" in b]
prox = pd.Series(data=[f[0]*1200 for f in prox], index=[f[1] for f in prox])
prox.sort_values(ascending=False).plot(kind='bar', figsize=(24, 10), fontsize=20)
```

Building attachment is very similar.

What conclusion can we draw from these comparisons? Well, city-assessed building value, it turns out, had a *much* weaker effect on our model than I would have thought; we've found that we were able to explain the majority of the variance without it, and that having it, though it does boost performance significantly, doesn't *structurally* change our model. This seems to show that the considerations made during the assessment process to a large extent weigh and take into account factors which have do not appear in the rest of our data (e.g. there is less-than-expected covariance between our variables and the city's assessed building value).

But how could that be? This really makes me wonder how they arrive at these (notoriously inaccurate, mind you) numbers in the first place...
