```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
train_test = pd.read_csv("../data/processed/likely_sales_featurized.csv", index_col=0)
```

Before we do anything more complicated, let's build a linear model on our data to see if it has any explanatory power.

First, let's build a model that ignores the city assessment values that we know. Can we independently get similar-ish performance?

```python
nonassessment_features = list(set(train_test.columns) - {'Borough', 'Block', 'Lot', 'SalePrice', 'ZipCode', 'Address', 
                                                         'AssessmentValueSqFt', 'EstPriorMarketValueSqFt',
                                                         'EstCurentMarketValueSqFt', 'MarketValueSqFt'})
target_variable = 'MarketValueSqFt'
```

```python
nonassessment_train_test = train_test[nonassessment_features]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
y = train_test['MarketValueSqFt']
clf.fit(nonassessment_train_test, y)
```

```python
y_pred = clf.predict(nonassessment_train_test)
```

```python
import sklearn.metrics
```

```python
mse = sklearn.metrics.mean_squared_error(y, y_pred)
mse
```

```python
root_mse = mse**(1/2)
root_mse
```

We're off by 81 \$/sq. ft., on average. This means that for an averagely-sized home of 1,200 square feet (the Manhattan average), we're off by \$103,200. This is actually pretty good!

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

Here is what this looks like when we plot this against the overall `MarketValueSqFt` curve:

```python
sns.kdeplot(y)
ax = plt.gca()
ax.set_xlim([0, 600])
plt.axvline(y.median(), c='black')
ax.add_patch(
    patches.Rectangle((y.median() - root_mse, 0), 2*root_mse, 0.006,
                      color='r', alpha=0.2)
)
```

$R^2$ is ~70%. This is not "amazing" by any means, but it's a respectable tally for such a simple attempt.

```python
r_squared = sklearn.metrics.r2_score(y, y_pred)
r_squared
```

Here are the variables, in order of value:

```python
coefs = list(zip(clf.coef_, nonassessment_features))
sorted(coefs)[::-1]
```

Parsing these is going to require piecing back together all of the one-hot encoded variables we've generated.

First of all, let's loot at fire companies, which encode *geographic location*. Naturally we expect this to be a very dominant variable in determining market price.

```python
firecomps = [(a, b) for a, b in coefs if "FireComp" in b]
firecomps = pd.Series(data=[f[0]*1200 for f in firecomps], index=[f[1] for f in firecomps])
```

Here are the price effects of the top 20 fire companies. To aid comprehension, I've multiplied their values by the sizes of the average Manhattan home (1200 square feet) to show the relative mangitude of their effects in those terms.

```python
firecomps.sort_values(ascending=False).head(20).plot(kind='bar', figsize=(24, 10), fontsize=26)
```

And the bottom 20:

```python
firecomps.sort_values(ascending=True).head(20).plot(kind='bar', figsize=(24, 10), fontsize=26)
```
